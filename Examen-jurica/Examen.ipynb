```python
import pandas as pd
df = pd.read_csv(r'csv/casas-pro.csv', dtype="string")
df1 = pd.read_csv(r'csv/departamentos-pro.csv', dtype="string")
#limpieza casas
df['precio'] = df['precio'].fillna('0')
df['precio'] = df['precio'].apply(lambda x: x.replace('$', '').replace(',', '').replace('MN','').replace(' ','').replace('NM','')).astype(float)
df['mets'] = df['mets'].fillna('0')
df['mets'] = df['mets'].apply(lambda x: x.replace(' ','').replace('m²', '')).astype(float)
#Limpieza departamentos
df1['precio'] = df1['precio'].fillna('0')
df1['precio'] = df1['precio'].apply(lambda x: x.replace('$', '').replace(',', '').replace('MN','').replace(' ','').replace('NM','')).astype(float)
df1['mets'] = df1['mets'].fillna('0')
df1['mets'] = df1['mets'].apply(lambda x: x.replace(' ','').replace('m²', '')).astype(float)
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
dfcasas = df[df['web-scraper-start-url'].str.contains("Jurica", na=False, case=False)]
dfcasas.reset_index(drop=True, inplace=True)
dfcasas.loc[:,['titulo','precio','habts','ban','mets']]
```

```python
#70% casas
casasFin = dfcasas.iloc[0:525]
metros = pd.DataFrame(casasFin['mets'])
precios = pd.DataFrame(casasFin['precio'])
model = linear_model.LinearRegression()
model.fit(metros, precios)
model.predict(pd.DataFrame([400]))
```

```python
#30% restante casas
casasFin1 = dfcasas.iloc[526:]
metros1 = pd.DataFrame(casasFin1['mets'])
precios1 = pd.DataFrame(casasFin1['precio'])
model = linear_model.LinearRegression()
model.fit(metros1, precios1)
model.predict(pd.DataFrame([500]))
```

```python
#departamentos
dfdeptos = df1[df1['web-scraper-start-url'].str.contains("Jurica", na=False, case=False)]
dfdeptos.reset_index(drop=True, inplace=True)
dfdeptos.loc[:,['titulo','precio','habts','ban','mets']]
```

```python
#70% deptos
deptosFin = dfdeptos.iloc[0:37]
metrosdeptos = pd.DataFrame(deptosFin['mets'])
preciosdeptos = pd.DataFrame(deptosFin['precio'])
modeldep = linear_model.LinearRegression()
modeldep.fit(metrosdeptos, preciosdeptos)
modeldep.predict(pd.DataFrame([100]))
```

```python
#30% deptos
deptosFin1 = dfdeptos.iloc[37:]
metrosdeptos1 = pd.DataFrame(deptosFin1['mets'])
preciosdeptos1 = pd.DataFrame(deptosFin1['precio'])
modeldep1 = linear_model.LinearRegression()
modeldep1.fit(metrosdeptos1, preciosdeptos1)
modeldep1.predict(pd.DataFrame([100]))
```
